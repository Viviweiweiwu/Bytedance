```python
all_data.loc[all_data.start_hour == 0, 'start_hour'] = 24
all_data.loc[all_data.end_hour == 0, 'end_hour'] = 24

all_data.loc[all_data.start_hour == 1, 'start_hour'] = 25
all_data.loc[all_data.end_hour == 1, 'end_hour'] = 25

all_data.loc[all_data.start_hour == 2, 'start_hour'] = 26
all_data.loc[all_data.end_hour == 2, 'end_hour'] = 26

all_data.loc[all_data.start_hour == 3, 'start_hour'] = 27
all_data.loc[all_data.end_hour == 3, 'end_hour'] = 27

all_data.loc[all_data.start_hour == 4, 'start_hour'] = 28
all_data.loc[all_data.end_hour == 4, 'end_hour'] = 28
```

```python
new_data = pd.DataFrame(columns = ["study_dt", "study_hour", "user_cnt"])
new_data
```

```python
for i in range(0, len(all_data)):
    sub_data = all_data.loc[i,]
    duration = sub_data.end_hour - sub_data.start_hour
    for j in range(0, duration+1):
        insert_study_dt = sub_data.study_dt
        insert_study_hour = sub_data.start_hour + j
        insert_user_cnt = sub_data._col3
        new_data = new_data.append({"study_dt": insert_study_dt, "study_hour": insert_study_hour, "user_cnt": insert_user_cnt}, ignore_index = True)
```

```python
new_data
```

```python
plot_data = new_data[new_data.study_dt >= '2020-07-20']
plot_data.groupby(['study_hour']).sum()
```
